**Prediction using Supervised ML**

**Problem statement: Predict the percentage of an student based on the no. of study hours.**


Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn import metrics
```

Importing dataset from the link

```python
dataset=pd.read_csv("http://bit.ly/w-data")
```

```python
dataset
```

Number of rows and coloumns present in the dataset :

```python
dataset.shape
```

check the info of data

```python
dataset.describe()
```

**Visualisation of the dataset :**

```python
# Plotting the distribution of scores
plt.scatter(dataset['Hours'], dataset['Scores'])
plt.title('Hours vs Score')
plt.xlabel('Studied Hours')
plt.ylabel('Score')
plt.show()
```

Preparing the data

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Training the Simple Linear Regression model on the Training set :

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Training complete.")
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_
# Plotting for the test data
plt.scatter(X, y,color="green")
plt.plot(X, line,color="red");
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')
plt.show()
```

### Predicting the result

```python
# Testing data - In Hours
print(X_test)
```

Predicted Test set results:

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

### Comparing Actual values with Predicted values :

```python
dataset = pd.DataFrame({'Actual Score': y_test, 'Predicted Score': y_pred})  
dataset
```

### **Let's predict the score for 9.25 hours/day**

```python
print('Score of student who studied for 9.25 hours per day is ', regressor.predict([[9.25]]))
```

### Evaluating the Model :

```python
#Checking the efficiency of model
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-Square:', metrics.r2_score(y_test, y_pred))
```
